```
from model import CPM2DPose
import torch
import numpy as np
import torchvision
import tqdm
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import cv2
import matplotlib.pyplot as plt
import math
import pickle

from glob import glob
import pandas

device = 'cuda:0'
num_joints = 21
```

```
def dataAugment(path):
  img_path = sorted(glob(path + 'rgb/*.jpg'))
  for i in tqdm.tqdm(range(len(img_path))):
    img = cv2.imread(img_path[i], cv2.IMREAD_COLOR)
    img_flip = cv2.flip(img, 0)

    num = img_path[i].split('.')[0].split('/')[-1]
    new_num = int(num) + len(img_path)
    new_path = path + 'rgb/' + str(new_num) + '.jpg'
    print(new_path)
    cv2.imwrite(new_path, img_flip)

    img_pkl = path + 'meta/' + str(num) + '.pkl'
    pkl = pandas.read_pickle(img_pkl)
    coords_2d = pkl['coords_2d']
    
    new_coord = []
    for i in range(num_joints):
      new_coord1 = np.float32(img.shape[0]) - coords_2d[i][0]
      new_coord2 = coords_2d[i][1]
      new_coord.append([new_coord1, new_coord2])
    
    dic = { 'coords_2d': np.array(new_coord), 'img': pkl['img']}

    new_pkl = path + 'meta/' + str(new_num) + '.pkl'
    print(new_pkl)

    output = open(new_pkl, 'wb')
    pickle.dump(dic, output)
    output.close()
```

```
root = '/content/gdrive/MyDrive/20171069_jooheekim_cv_ass2/(CSE48001)Obman_dataset/'# change to relative path
train_path = root + 'train/'
dataAugment(train_path)
test_path = root + 'test/'
dataAugment(test_path)
```

```
class ObmanDataset(Dataset):
    def __init__(self, method=None):
        self.root = '/content/gdrive/MyDrive/20171069_jooheekim_cv_ass2/(CSE48001)Obman_dataset/' # The path to access data stored directory
        self.x_data = []
        self.y_data = []
        if method == 'train':
            self.root = self.root + 'train/'
            self.img_path = sorted(glob(self.root + 'rgb/*.jpg'))

        elif method == 'test':
            self.root = self.root + 'test/'
            self.img_path = sorted(glob(self.root + 'rgb/*.jpg'))

        for i in tqdm.tqdm(range(len(self.img_path))):
            img = cv2.imread(self.img_path[i], cv2.IMREAD_COLOR)
            print(self.img_path[i])
            b, g, r = cv2.split(img)
            img = cv2.merge([r, g, b])
            self.x_data.append(img)

            num = self.img_path[i].split('.')[0].split('/')[-1]
            img_pkl = self.root + 'meta/' + str(num) + '.pkl'
            pkl = pandas.read_pickle(img_pkl)
            coords_2d = pkl['coords_2d']
            self.y_data.append(coords_2d)

    def __len__(self):
        return len(self.img_path)

    def __getitem__(self, idx):
        transform1 = torchvision.transforms.ToTensor()
        new_x_data = transform1(self.x_data[idx])

        return new_x_data, self.y_data[idx], idx
```

```
class Trainer(object):
    def __init__(self, epochs, batch_size, lr):
        self.epochs = epochs
        self.batch_size = batch_size
        self.learning_rate = lr
        self._build_model()
        self.loss_func = torch.nn.MSELoss() # use MSELoss() as loss function
        self.optimizer = torch.optim.SGD(self.poseNet.parameters(), lr) # use SGD() as optimizer
        self.losses = [] # list to save the loss for each batch

        dataset = ObmanDataset(method='train')
        self.root = dataset.root
        self.dataloader = DataLoader(dataset, batch_size=self.batch_size, shuffle=True)

        # Load of pretrained_weight file
        weight_root = self.root.split('/')
        del weight_root[-2]
        weight_root = "/".join(weight_root)
        weight_PATH = weight_root+'pretrained_model/pretrained_weight.pth' # weight.pth should be in the pretrained_model directory
        # Please use this relative path in the local environment, but the pth file should in the same path as this ipynb
        # weight_PATH = './pretrained_weight.pth'
        self.poseNet.load_state_dict(torch.load(weight_PATH))

        print("Training...")

    def _build_model(self):
        # 2d pose estimator
        poseNet = CPM2DPose()
        self.poseNet = poseNet.to(device)
        self.poseNet.train()

        print('Finish build model.')

    def skeleton2heatmap(self, _heatmap, keypoint_targets):
        heatmap_gt = torch.zeros_like(_heatmap, device=_heatmap.device)

        keypoint_targets = (((keypoint_targets)) // 8)
        for i in range(keypoint_targets.shape[0]):
            for j in range(21):
                x = int(keypoint_targets[i, j, 0])
                y = int(keypoint_targets[i, j, 1])
                if x > 31: 
                  x = 31
                if y > 31:
                  y = 31
                heatmap_gt[i, j, x, y] = 1

        heatmap_gt = heatmap_gt.detach().cpu().numpy()
        for i in range(keypoint_targets.shape[0]):
            for j in range(21):
                heatmap_gt[i, j, :, :] = cv2.GaussianBlur(heatmap_gt[i, j, :, :], ksize=(3, 3), sigmaX=2, sigmaY=2) * 9 / 1.1772
        heatmap_gt = torch.FloatTensor(heatmap_gt).to(device)
        return heatmap_gt

    def train(self):
        date = '201105'
        for epoch in tqdm.tqdm(range(self.epochs + 1)):
            # if epoch % 2 == 0:
            #     torch.save(self.poseNet.state_dict(), "_".join([self.root+'trained_model/', date, str(epoch), 'model.pth']))

            running_loss = 0.0
            for batch_idx, samples in enumerate(self.dataloader):
                x_train, y_train = samples
                heatmapsPoseNet = self.poseNet(x_train.cuda())
                gt_heatmap = self.skeleton2heatmap(heatmapsPoseNet, y_train)
                loss = self.loss_func(heatmapsPoseNet, gt_heatmap)
                loss.backward()
                self.optimizer.step()
                self.losses.append(loss.item())

                running_loss += loss.item() # update running loss for intermediate checking purpose
                ## Write train result
                if batch_idx % 20 == 0:
                    with open('train_result_' + date + '.txt', 'a') as f:
                        f.write('Epoch {:4d}/{} Batch {}/{}\n'.format(
                            epoch, self.epochs, batch_idx, len(self.dataloader)
                        ))
                    print('Epoch {:4d}/{} Batch {}/{} Loss: {:3f}'.format(
                        epoch, self.epochs, batch_idx, len(self.dataloader), running_loss / 20
                    ))
                    running_loss = 0.0
        
        # model_path in Colab environment
        model_path = self.root.split('/')
        del model_path[-2]
        del model_path[-2]
        model_path = "/".join(model_path) + 'dataAUGweight.pth'
        # relative model_path is as belows:
        # model_path = './finetunedweight.pth'
        torch.save(self.poseNet.state_dict(), model_path)
        print('Finish training.')
```

```
epochs = 20
batchSize = 16
learningRate = 0.01

trainer = Trainer(epochs, batchSize, learningRate)
trainer.train()
```

```
class Tester(object):
    def __init__(self, batch_size):
        self.batch_size = batch_size
        self._build_model()

        dataset = ObmanDataset(method='test')
        self.root = dataset.root
        self.dataloader = DataLoader(dataset, batch_size=self.batch_size, shuffle=False)
        self.datalen = dataset.__len__()
        self.mse_all_img = []

        # Load of pretrained_weight file in Colab environment
        weight_root = self.root.split('/')
        del weight_root[-2]
        del weight_root[-2]
        weight_root = "/".join(weight_root)
        weight_PATH = weight_root+'dataAUGweight.pth'
        # Please use this relative path in the local environment, but the pth file should in the same path as this ipynb
        # weight_PATH = './finetunedweight.pth'
        self.poseNet.load_state_dict(torch.load(weight_PATH))

        print("Testing...")

    def _build_model(self):
        # 2d pose estimator
        poseNet = CPM2DPose()
        self.poseNet = poseNet.to(device)

        print('Finish build model.')

    def heatmap2skeleton(self, heatmapsPoseNet):
        skeletons = np.zeros((heatmapsPoseNet.shape[0], heatmapsPoseNet.shape[1], 2))
        for m in range(heatmapsPoseNet.shape[0]):
            for i in range(heatmapsPoseNet.shape[1]):
                u, v = np.unravel_index(np.argmax(heatmapsPoseNet[m][i]), (32, 32))
                skeletons[m, i, 0] = u * 8
                skeletons[m, i, 1] = v * 8
        return skeletons

    def test(self):
        for batch_idx, samples in enumerate(self.dataloader):
            x_test, y_test, idx = samples
            heatmapsPoseNet = self.poseNet(x_test.cuda()).cpu().detach().numpy()
            skeletons_in = self.heatmap2skeleton(heatmapsPoseNet)
            for i in range(len(idx)):
              error = self.calc_error(y_test[i], skeletons_in[i])
              self.mse_all_img.append(error)
        self.print_error()
    
    def calc_error(self, true, pred):
        err = 0
        for i in range(num_joints):
              y_pred = pred[i][0]
              x_pred = pred[i][1]
              y_true = true[i][0].item()
              x_true = true[i][1].item()
              err += math.sqrt((x_pred - x_true)**2 + (y_pred - y_true)**2)
        err = err / num_joints
        return err
    
    def print_error(self):
      err = 0
      num = len(self.mse_all_img)
      for i in range(num):
        err += self.mse_all_img[i]
      err = err / num
      print('Average error is ' + str(err))
```
